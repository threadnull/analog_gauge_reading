```python
import cv2
import numpy as np
```

```python
niddle = cv2.imread("image/pressure_gauge/niddle.png")
gauge = cv2.imread("image/pressure_gauge/pressure_gauge.png")
```

```python
def get_center(img):
    return (img.shape[1]//2, img.shape[0]//2)
```

```python
def rotate_img(img, angle, scale=1):
    rotation_matrix = cv2.getRotationMatrix2D(get_center(img), angle, scale)
    rotated_img = cv2.warpAffine(img, rotation_matrix, (img.shape[1], img.shape[0]))
    return rotated_img
```

```python
cv2.imshow("niddle", rotate_img(niddle, -180))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
