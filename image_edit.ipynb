```python
import cv2
import numpy as np
```

```python
niddle = cv2.imread("image/pressure_gauge/niddle.png")
gauge = cv2.imread("image/pressure_gauge/pressure_gauge.png")
```

```python
def get_center(img):
    return (img.shape[1]//2, img.shape[0]//2)
```

```python
angle = -180
scale = 1
rotation_matrix = cv2.getRotationMatrix2D(get_center(niddle), angle, scale)
rotated_img = cv2.warpAffine(niddle, rotation_matrix, (niddle.shape[1], niddle.shape[0]))
```

```python
cv2.imshow("niddle", rotated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
