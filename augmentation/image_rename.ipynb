```python
import cv2
import os
```

```python
input_path = "..\\image\\temp"
output_path = "..\\image\\rename"
image_list = os.listdir(input_path)
print(image_list)
```

```python
image_path = []
for file in image_list:
    file_path = os.path.join(input_path, file)
    image_path.append(file_path)
print(image_path)
print(len(image_path))
```

```python
cnt = 0
for path in image_path:
    # file = os.path.split(path)[-1]
    image = cv2.imread(path)
    save_path = os.path.join(output_path, f"augmented_gauge_{str(cnt)}.jpg")
    print(save_path)
    cv2.imwrite(save_path, image)
    cnt += 1
print("rename done")
```
