```python
import cv2
import os
```

```python
def get_center(img):
    return (img.shape[0] // 2, img.shape[1] // 2)
```

```python
center = (1050, 540)

def get_crop_with_resize(img):
    crop_size = 1080, 1080
    # center = get_center(img)

    h_start = center[0] - (crop_size[0] // 2)
    h_end = h_start + crop_size[0]
    w_start = center[1] - (crop_size[1] // 2)
    w_end = w_start + crop_size[1]

    cropped_image = img[h_start:h_end, w_start:w_end]
    resized_image = cv2.resize(cropped_image, (640, 640))

    return resized_image
```

```python
path = "capture"
image_list = os.listdir(path)
print(image_list)
```

```python
image_path = []
for file in image_list:
    file_path = os.path.join(path, file)
    image_path.append(file_path)
print(image_path)
```

```python
for path in image_path:
    file = os.path.split(path)[-1]
    image = cv2.imread(path)
    resized_image = get_crop_with_resize(image)
    cv2.imwrite(f"resize\\re_{file}", resized_image)
    # break
```
