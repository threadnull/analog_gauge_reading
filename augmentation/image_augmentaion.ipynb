```python
import os
import cv2
import albumentations as A
```

```python
input_image = "..\\image\\origin"
output_imgae = "..\\image\\augmented"
# 증강할 이미지 갯수
num_aug_per_img = 5
```

```python
transform = A.Compose([
    # 회전
    A.SafeRotate(limit=180, p=0.8, border_mode=cv2.BORDER_CONSTANT),
    A.RandomScale(scale_limit=0.2, p=0.5),

    # 노이즈
    A.OneOf([
        A.GaussNoise(var_limit=(10.0, 50.0), p=1.0), 
        A.MultiplicativeNoise(multiplier=(0.9, 1.1), p=1.0), 
        A.ISONoise(color_shift=(0.01, 0.05), intensity=(0.1, 0.5), p=1.0)
    ], p=0.6),

    # 화질
    A.OneOf([
        A.GaussianBlur(blur_limit=(3, 7), p=1.0),
        A.MotionBlur(blur_limit=(3, 7), p=1.0),
        A.RandomRain(brightness_coefficient=0.9, drop_length=20, p=1.0),
    ], p=0.3),

    # 조명
    A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.7),
    A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=30, val_shift_limit=20, p=0.5),
    A.RandomGamma(gamma_limit=(80, 120), p=0.5),

    # 이미지 크기
    A.PadIfNeeded(min_height=640, min_width=640, border_mode=cv2.BORDER_CONSTANT),
    A.CenterCrop(height=640, width=640, p=1.0),
])
```

```python
image_list = []
path = "..\\image\\origin"
for img in os.listdir(path):
    image_list.append(os.path.join(path, img))
print(len(image_list))
```

```python
def image_augmentation():
    for img_path in image_list:
        file_name = os.path.basename(img_path)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        base_name = os.path.splitext(file_name)[0]

        for i in range(num_aug_per_img):
            augmented = transform(image=img)
            aug_img = augmented["image"]

            out_name = f"{base_name}_augmented_{i:03d}.jpg"
            save_path = os.path.join(output_imgae, out_name)
            
            cv2.imwrite(save_path, cv2.cvtColor(aug_img, cv2.COLOR_RGB2BGR))
    print("augmentaion done")
```

```python
image_augmentation()
```
