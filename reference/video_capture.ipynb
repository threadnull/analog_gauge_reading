```python
import cv2
```

```python
vidcap=cv2.VideoCapture("video/gauge.mp4")

count = 0
ret=True
while(ret):
    ret, image = vidcap.read()#return 값과 image를 읽어옴
    if(ret==False):
        break

    if(int(vidcap.get(1)) % 5 == 0):#50 frame당 한 frame만 저장
        print('Saved frame number: '+str(int(vidcap.get(1))))
        cv2.imwrite("capture/frame%d.jpg" % count, image)#새롭게 .jpg 파일로 저장
        print('Saved frame%d.jpg' % count)
        count += 1
vidcap.release()
```
